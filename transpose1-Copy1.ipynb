```python
import pandas as pd;
df = pd.DataFrame([['Month','Jan','Feb'],['Facebook', 200, 220],['Adwords',150,112]],columns=list('ABC'));
df.head()
```

```python
firstCol = df.columns.tolist()[0];
print firstCol
```

```python
df = df.set_index(df[firstCol]);
df.head()
```

```python
df = df.drop(firstCol, 1);
df.head()
```

```python
dft = df.T
df.head()
```

```python
dft[firstCol] = dft.index;
dft.head()
```

```python
cols = dft.columns.tolist()
cols = cols[-1:] + cols[:-1]
dft = dft[cols]
```

```python
dft.head()
```

```python
result = pd.DataFrame(dft.values.tolist(),columns=dft.columns.tolist());
```
